<h1 style="color:#FF69B4"><li>Training the Xception model </li></h1>
<h1 style="color:#FF69B4"><li> Load model </li></h1>

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.xception import Xception
import os


train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
        '/content/drive/MyDrive/GarbageClassification/GarbageData/train_data',
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical',
        classes=['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash'])


base_model = Xception(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)
x = tf.keras.layers.Dense(1024, activation='relu')(x)
predictions = tf.keras.layers.Dense(6, activation='softmax')(x)
model = tf.keras.models.Model(inputs=base_model.input, outputs=predictions)


for layer in base_model.layers:
    layer.trainable = False


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


model.fit(
    train_generator,
    steps_per_epoch=train_generator.n // train_generator.batch_size,
    epochs=10)


model.save('my_model.h5')
```

<h1 style="color:#FF69B4"><li> Predict the image with the difference classes and Accuracy between 0 - 100% </li></h1>

```python

model = tf.keras.models.load_model('/content/drive/MyDrive/GarbageClassification/Data/my_model.h5')


img_paths = ['/content/drive/MyDrive/GarbageClassification/Data/cardboard/cardboard151.jpg',
             '/content/drive/MyDrive/GarbageClassification/Data/glass/glass161.jpg',
             '/content/drive/MyDrive/GarbageClassification/Data/metal/metal258.jpg',
             '/content/drive/MyDrive/GarbageClassification/Data/paper/paper259.jpg',
             '/content/drive/MyDrive/GarbageClassification/Data/trash/trash34.jpg',
             '/content/drive/MyDrive/GarbageClassification/Data/plastic/plastic359.jpg']

images = []
for path in img_paths:
    img = image.load_img(path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array /= 255
    images.append(img_array)

images = np.stack(images)

preds = model.predict(images)


classes = ['cardboard', 'glass', 'metal', 'paper', 'plastic', 'trash']
for i in range(6):    
    predicted_class = classes[np.argmax(preds[i])]
    predicted_accuracy = round(preds[i][np.argmax(preds[i])] * 100, 2)
    print(f"Predicted class of image {i+1}: {predicted_class}")
    print(f"Predicted accuracy of image {i+1}: {predicted_accuracy}%")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the input shape of the images
input_shape = (224, 224, 3)

# Create an instance of ImageDataGenerator for test data
test_datagen = ImageDataGenerator(rescale=1./255)

# Load the test images from the directory
test_generator = test_datagen.flow_from_directory(
    '/content/drive/MyDrive/GarbageClassification/GarbageData/test_data',
    target_size=input_shape[:2],
    batch_size=1,
    class_mode='categorical',
    shuffle=False
)

# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(test_generator)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
import numpy as np
from sklearn.metrics import classification_report

# Predict the class labels for the test data
test_predictions = model.predict(test_generator)
predicted_classes = np.argmax(test_predictions, axis=1)

# Get the true class labels from the test data generator
true_classes = test_generator.classes

# Get the class labels (names) from the generator's class indices
class_labels = list(test_generator.class_indices.keys())

# Generate a classification report
classification_rep = classification_report(true_classes, predicted_classes, target_names=class_labels)

print("Classification Report:")
print(classification_rep)

```
